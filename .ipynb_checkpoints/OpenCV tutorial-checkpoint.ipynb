```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

img = cv.imread("./data/Well B1_01/2020_02_21_19_01_45_57.tif")
plt.imshow(img)
plt.show()
```

```python
img = cv.cvtColor(img, cv.CV_8U)
print(img.shape)
```

```python
kernel = np.array([[0, -1, 0], 
                   [-1, 5, 1],
                   [0, -1, 0]], np.float32)
dst1 = cv.filter2D(img, -1, kernel)
plt.imshow(dst1)
```
